# The beginning of quantum mechanics. Or: Trouble in paradise

Starting with Newton and Kepler, then Maxwell, Boltzmann, Gibbs, etc, classical physics explained a whole lot of stuff: Motion of planets, electricity and magnetism, thermodynamics. But there were problems appearing around the turn of the (last) century:

1. Blackbody radiation: Hot objects emit electromagnetic radiation, but not uniformly across all wavelengths. Classical physics worked well for long wavelengths/low energy, but failed badly at the ultraviolet end. 

2. Photoelectric effect: Metals eject electrons when you shine light on them, but the maximum energy of the ejected electrons depnds on the frequency of the light, not its intensity. At the time it was taken for granted that higher intensity beams should have higher energy electrons.

3. The Rutherford atom and the plum pudding model: The Thomson model for solid materials was a fairly uniform "plum pudding" mixture of postive and negative bits. But, when it became possible to hit a thin gold film with a beam of particles like neutrons, they sometimes bounced straight back! Rutherford proposed that all the positive charge concentrated in a tiny region, with electrons orbiting. But why didn't they fall inward?

4. Atoms were found to emit/absorb light at specific, discrete frequencies. These series of lines were catalogued (Balmer, LYMAN, etc), but there was no explanation for why an atom could only emit at discrete wavelengths. 

The beginnings of an answer started in the first years of the 1900's:

1. Planck explained the blackbody spectrum with the assumption that light could only be emitted at discrete frequencies: $E = h\nu$. $h$ of course is now called Planck's constant. No energy could be emitted from a blackbody for $E > kT$. 

2. The quantum nature of light was then used by Einstein to explain the photoelectric effect, suggesting the wave-particle duality. 

3. The Bohr model for the atom explained the spectral series: A bound atomic system has only discrete energy states, with $E_n = -13.6 eV/n^2$. Therefore a transition between two states had to involve absorption/emission of a photon with a wavelength $\nu = (E_1 - E_2)/h$.

4. Then of course, the double slit experiment. 

## Quantum Mechanics vs Classical Mechanics

Consider a particle of mass $m$ in a potential $V$. The total energy of this (conservative) system is $H(\mathbf{r},\mathbf{p}) = \mathbf{p}^2/2m + V(\mathbf{r})$. 
* In classical mechanics, the particle has an exact position $\mathbf{r}$ and momentum $\mathbf{p}$, and its motion is deterministic. 
* In quantum mechanics, we can only know the product of $\delta x$ and $\delta p_x$ to within the limit given by Heisenberg: $\delta x \delta p_x > \hbar/2$. As a result, qm particles are not deterministic, and we can only discuss the *probability* of finding them somewhere with some momentum.

QM particles are described by wavefunctions $\psi(\mathbf{r})$, which are complex functions of space. If

$$\psi = a + ib$$

then the complex conjugate we denote with a $\dagger$:

$$\psi^{\dagger} = a - ib$$ 

and the product 

$$\psi^{\dagger}\psi = a^2 + b^2$$ is real and positive definite. 

It turns out that $\psi^{\dagger}\psi$ is the probability density to find the particle within a region $d^3r$. This is called the Born rule. (If you want to go down a fun rabbit hole, see if you can find an explanation for why this should be so. You might start with this paper https://arxiv.org/abs/quant-ph/0401062 , which has the added bonus of beginning with the line "This recreational paper..." I want to write a "recreational paper"!) 

This means that 

$$P(\mathbf{r})d^3r = \psi^{\dagger}\psi d^3r$$ 

and that if we integrate it over all space, it must equal $1$ because it is a probability:

$$\int_{-\infty}^{+\infty}\psi^{\dagger}\psi d^3r = 1 \tag{1}$$

This is how we compute normalization coefficients.

## Time-dependence: The Schrodinger Equation

The time-dependence of $\psi$ is governed by the Schrodinger equation (SE) in QM:

$$i\hbar\frac{\partial \psi}{\partial t} = -\frac{\hbar^2}{2m}\nabla^2\psi(\mathbf{r},t) + V(\mathbf{r})\psi(\mathbf{r},t)\tag{2}$$

To develop some intuition about QM versus CM, let's consider a 1D system with $V=0$. Clasically, if $V=0$, $F=0$ and therefore $p$ is constant. We call this a "free particle." The 1D SE for a free particle reads:

$$\frac{\partial \psi}{\partial t} = \frac{i\hbar}{2m}\frac{\partial^2}{\partial x^2}\psi(\mathbf{r},t)\tag{3}$$

Consider a WF which is initially localized --- maybe it's a Gaussian. **Draw this on the board...too lazy to make a figure** Consider the RHS of Eq. 3: Are there regions where the 2nd derivative is positive? Negative? Based on this, what would you infer about $\partial \psi/\partial t$ in these regions? (Recall also the comparison to a diffusion equation.)

Let's use a scaling argument to see how fast the WF spreads. Look at Eq. 3, but instead of the partial derivatives, consider little finite differences:

$$\begin{align} \frac{\Delta\psi}{\Delta t} & \approx \frac{\hbar}{2m}\frac{\Delta\psi}{\Delta x^2} \tag{4} \\
\Delta t  & \approx \frac{2m}{\hbar}(\Delta x)^2 \tag{5} \\
\end{align}$$

Several things are implied by this relation:
1. Larger $\hbar$, slower spread
2. Larger $m$, slower spread
3. More initial confinement, faster spread.

Consider: 
* A 1 kg mass, confined to a region of 1 m. $\Delta t \approx 10^{26}$ years. QM plays no role.
* Electron in a hydrogen atom. $m \approx 10^{-31}$ kg, $\Delta x \approx 10^{-10}$. $\Delta t \approx 10^{-17}$ sec. QM matters, big time.

## The terms of the SE: A fast and not-very-deep tour

Since particles apparently have some wave-like character (see double slit experiment), let's use a plane-wave solution to the S.E. to think a bit more about what each term means. Let:

$$\psi(x,t) = \psi_0e^{i(kx - \omega t)}$$

If we differentiate w.r.t. time:

$$\begin{align}
\frac{\partial \psi}{\partial t} & = -i\omega \psi(x,t) \\
i\hbar \frac{\partial \psi}{\partial t} & = \hbar\omega\psi(x,t) \tag{6}\\
\end{align}$$

**Question:** What are the units of $\hbar\omega$? Apparently then, 

$$i\hbar \frac{\partial}{\partial t} \Leftrightarrow E \tag{7}$$ 

Operate on $\psi$ with $i\hbar \partial/\partial t$, get back the energy. Now consider the spatial derivative:

$$\begin{align}
\frac{\partial \psi}{\partial x} & = ik\psi(x,t) \\
-i\hbar\frac{\partial \psi}{\partial x} & = \hbar k \psi(x,t) \tag{8}\\
\end{align}$$

**Question:** What are the units of $\hbar k$? Apparently, then: 

$$-i\hbar \frac{\partial}{\partial x} \Leftrightarrow P \tag{9}$$

Operate on $\psi$ with $-i\hbar \partial/\partial x$, get back the momentum.

## Observables and expectation values

According to the Copenhagen interpretation of QM, we can only ever predict the *probability* of finding a quantum particle at a particular position, or with a particular momentum, or in a state with a particular energy. We compute observables in QM via "expectation values." For example (assuming a normalized $\psi$), the expected location of a particle is found by:

$$\langle x \rangle = \int_{-\infty}^{+\infty}\psi^{\dagger}x\psi \tag{10}$$

We can use the energy and momentum operators just introduced to write:

$$\begin{align}
\langle E \rangle & = \int_{-\infty}^{+\infty}\psi^{\dagger}\left(i\hbar\frac{\partial}{\partial t}\right)\psi \tag{11} \\
\langle p \rangle & = \int_{-\infty}^{+\infty}\psi^{\dagger}\left(-i\hbar\frac{\partial}{\partial x}\right)\psi \tag{12} \\
\end{align}$$

Looking back at the S.E. we define 

$$\hat{H} = -\frac{\hbar^2}{2m}\nabla^2 + V \tag{13}$$

which means that we can write the S.E. compactly as

$$i\hbar\frac{\partial \psi}{\partial t} = \hat{H}\psi \tag{14}$$ 

and the energy expectation value as

$$\langle E \rangle = \int_{-\infty}^{+\infty}\psi^{\dagger}\hat{H}\psi \tag{15}$$
