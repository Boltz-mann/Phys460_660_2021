## Solving the Schroedinger Equation, Old Skool

The Schroedinger equation (S.E. hereafter for short) is a PDE, which makes life difficult. The general approach to attempt an exact solution (which is not possible in most cases!) is to:

1. Separate the variables (time and space).
2. Solve each part separately.
3. Impose boundary conditions to find the (quantized) stationary states.
4. The solution for any initial condition is expressed as a sum of stationary states.

First we will go through these steps formally, then apply them to a classic problem --- the particle in a box.

### Separation of Variables

A standard technique for solving PDE's for the S.E., let

$$\psi(\mathbf{r},t) = R(\mathbf{r})T(t)  \tag{Eq. 1}$$ 

Then substitute our ansatz into the S.E.:

$$\begin{align}
i\hbar\frac{\partial\psi(\mathbf{r},t)}{\partial t} &= \frac{-\hbar^2}{2m}\nabla^2\psi(\mathbf{r},t) + V(\mathbf{r})\psi(\mathbf{r},t) \tag{Eq. 2} \\
i\hbar\frac{\partial}{\partial t}\left[ R(\mathbf{r})T(t)\right ] &= \frac{-\hbar^2}{2m}\nabla^2 \left[R(\mathbf{r})T(t)\right ] + V(\mathbf{r})\left [R(\mathbf{r})T(t)\right ] \tag{Eq. 3}\\
i\hbar\frac{\partial T(t)}{\partial t} R(\mathbf{r}) &=  \frac{-\hbar^2}{2m}\left[\nabla^2 R(\mathbf{r}) \right] T(t) + V(\mathbf{r})R(\mathbf{r})T(t) \tag{Eq. 4}\\ 
\end{align}$$

Now divide through both sides by $R(\mathbf{r})T(t)$:

$$i\hbar\frac{1}{T}\frac{\partial T(t)}{\partial t} = \frac{-\hbar^2}{2m}\frac{1}{R}\nabla^2 R(\mathbf{r}) + V(\mathbf{r}) \tag{Eq. 5}$$

### Time Dependence
Now we note that the L.H.S. of Eq. 5 is a function of $t$ only and the R.H.S. is only a function of $r$. That means they must both be equal to the same constant. Let's call it $E$ (foreshadowing for what lies ahead). We can immediately integrate the L.H.S.:

$$\begin{align}
i\hbar\frac{1}{T}\frac{\partial T}{\partial t} &= E \\
\frac{\partial T}{\partial t} &= \frac{-i}{\hbar}ET \\
T(t) &= Ae^{\frac{-iEt}{\hbar}} \tag{Eq. 6}\\
\end{align}$$

### Spatial Dependence
Aha! The time dependence of $\psi$ is only a function of $E$. Now let's look at the spatial dependence. Returning to Eq. 5, but now looking at the R.H.S. (remember: both sides of Eq. 5 must equal the same constant, which we called "$E$").

$$\begin{align}
\frac{-\hbar^2}{2m}\frac{1}{R}\nabla^2 R + V(\mathbf{r}) &= E \\
\frac{-\hbar^2}{2m}\nabla^2 R + V(\mathbf{r})R &= ER\\
\hat{H}R &= ER \tag{Eq. 7}\\
\end{align}$$

where as in the last notebook we define

$$\hat{H} = \frac{-\hbar^2}{2m}\nabla^2 + V(\mathbf{r})  \tag{Eq. 8}$$

Note that since we have separated the spatial and time dependence in $\psi$, and since $\hat{H}$ only depends on the spatial variables, we can also write:

$$\hat{H}\psi = E\psi \tag{Eq. 9}$$

This is a matrix equation: The matrix $\hat{H}$ multiplies a vector $\psi$, and returns a scalar $E$ times the same vector $\psi$. In other words, $E$ is an eigenvalue of $\hat{H}$ and $\psi$ is the corresponding eigenvector, as is the purely spatial part of the eigenvector $R$.

Because there is no time-dependence in Eq. 7, the eigenvectors $R$ represent *stationary states* of $\hat{H}$. They are discrete, and we will index them by an (integer) quantum number $\ell$:

$$\begin{align}
\hat{H}R_1 &= E_1 R_1 \\
\hat{H}R_2 &= E_2 R_2 \\
\vdots \\
\hat{H}R_{\ell} &= E_{\ell} R_{\ell} \tag{Eq. 10}\\
\end{align}$$

The stationary states and eigenvalues are determined by the boundary conditions and the potential $V$. (I'll remind you of how this works in an example below.)

What about the full time dependent solution? As usual when solving for the time dependent solution of a differential equation we need an initial condition: $\psi(r,t=0)$. Because the eigenvectors form a complete basis, we can write the initial condition as a sum of the eigenvectors, wirth appropriately chosen amplitudes $A_{\ell}$:

$$\psi(r,t=0) = \sum_{\ell=1}^{\infty}A_{\ell}R_{\ell}(r) \tag{Eq. 11}$$

This is a projection, just like the projection of the dynamics of a system of coupled oscillators onto the normal modes. In this case, however, the eigenstates that form a basis for the preojection need not be sines and cosines â€” they are determined by the BC and potential. So if $V$ is the infinite well, you get cosines for $R_{\ell}$; if $V$ is the harmonic oscillator you get Hermite polynomials, etc.

### Full Solution
Since the solution is separable and we have already obtained the time-dependent solution, we can immediately write down a formal solution for $\psi$ at any time $t$:

$$\psi(r,t) = \sum_{\ell=1}^{\infty}A_{\ell}R_{\ell}(r)e^{-iE_{\ell}t/\hbar} \tag{Eq. 12}$$

### Projection and Orthonormality

The amplitudes $A_{\ell}$ are determined using the orthonormal property of the basis, just like you would do when determining discrete Fourier coefficients, or normal mode amplitudes, or in any other problem in which the solution is expressed in the form of Eq. 11. Orthonormality means

$$\int d^3rR_{\ell}R_n = \begin{cases}
1 & \text{if } \ell = n \\
0 & \text{if } \ell \neq n 
\end{cases} \tag{Eq. 13}$$

We can use this property to derive an expression for the amplitudes $A_{\ell}$ by starting with our expression for $\psi{r,t=0}$ in Eq. 11, then multiplying through by $R_n$ and then integrating to project onto the basis:

$$\begin{align}
\psi(r,t=0) &= \sum_{\ell=1}^{\infty}A_{\ell}R_{\ell}(r) \\
R_n(r)\psi(r,t=0) &= \sum_{\ell=1}^{\infty}A_{\ell}R_n(r)R_{\ell}(r) \\
\int d^3r R_n(r)\psi(r,t=0) &= \sum_{\ell=1}^{\infty}A_{\ell}\int d^3r R_n(r)R_{\ell}(r) \\
\end{align}$$

The integral on the RHS is zero unless $n=\ell$, which gives

$$A_{\ell} = \int d^3r R_{\ell}(r)\psi(r,t=0) \tag{Eq. 14}$$

### Particle in a box

Consider a particle confined in one dimension by an infinite square well, as shown in the figure. 

<img src="./PinBox.jpeg" alt="square well" width="200"/>

In regions I and III, $V(x) = \infty$ and in region II $V(x) = 0$. We start by obtaining the stationary solutions in region II. There we have the time independent SE with V=0:

$$\begin{align}
\hat{H}R & = ER \\
\frac{-\hbar^2}{2m}\frac{\partial ^2 R}{\partial x^2} &= ER \\
\frac{\partial ^2 R}{\partial x^2} &= \frac{-2mE}{\hbar^2}R \tag{Eq. 15}\\
\end{align}$$

This is the wave equation, with the general solution

$$R(x) = A\text{sin}\left(\sqrt{\frac{2mE}{\hbar^2}}x \right) + B\text{cos}\left(\sqrt{\frac{2mE}{\hbar^2}}x \right) \tag{Eq. 16}$$

Now we use the boundary conditions to find A and B. $R(x=0) = 0$ since the potential is infinity there. This means tehre are no cosine solutions, and so $B=0$. We also have $R(x=L)=0$, which means that the sin function must have a node at $x=L$: 

$$A\text{sin}\left(\sqrt{\frac{2mE}{\hbar^2}}L \right) = 0$$

which is satisfied when

$$\sqrt{\frac{2mE}{\hbar^2}}L = n\pi$$

with $n$ taking positive integer values: $n=1,2,3,...$ Solving for E we obtain:

$$E_n = \frac{n^2\pi^2\hbar^2}{2mL^2} \tag{Eq. 17}$$

Note that the allowed energies are *discretized* and indexed by $n$, which is our quantum number for the particleinabox. 

Any solution (ie, wavefunction) which satisfies the BC can be written as a sum of sines, with the energies discrtized as in Eq. 17:

$$\begin{align}
R(x) &= \sum_{n=1}^{\infty}A_n \text{sin}\left(\sqrt{\frac{2mE_n}{\hbar^2}}x\right) \\
&=\sum_{n=1}^{\infty}A_n \text{sin}(k_n x) \tag{Eq. 18}\\
\end{align}$$

where in the last line we defined 

$$k_n = \sqrt{\frac{2mE_n}{\hbar^2}}$$

The requirement that the wavefunction be normalized permits detrmination of the coefficients $A_n$:

$$\begin{align}
1 &= \int_{-\infty}^{+\infty} R^{\dagger}(x)R(x)dx \\
&= A_n^2 \int_{-\infty}^{+\infty}\text{sin}^2(k_nx)dx \\
&= A_n^2 \int_{-\infty}^{+\infty}\text{sin}^2\left(\frac{n\pi}{L}x \right)dx \\
\end{align}$$

Which we solve for $A_n$ to find that $A_n = \sqrt{2/L}$ for all $n$. We now have the stationary solutions, indexed by $n$:

$$R_n(x) = \sqrt{\frac{2}{L}}\text{sin}(k_nx)$$

The lowest energy ground state ($n=1$) has $E_1 = \frac{\pi^2\hbar^2}{2mL^2}$ with a wavefunction with one node at either end. The first excited state ($n=2$) has $E_2 = \frac{4\pi^2\hbar^2}{2mL^2}$ and a wavefunction with a node in the middle, and so on.
